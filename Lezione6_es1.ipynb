### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# 1. Genera 1000 valori dalla distribuzione gamma con parametro di forma 1
shape_param = 1
sample = gamma.rvs(shape_param, size=1000)

# 2. Istogramma + PDF teorica
x = np.linspace(0, max(sample), 1000)
pdf = gamma.pdf(x, shape_param)

plt.figure(figsize=(10, 5))
plt.hist(sample, bins=30, density=True, alpha=0.6, label='Campione')
plt.plot(x, pdf, 'r-', label='PDF teorica (gamma(1))')
plt.title("Distribuzione Gamma (forma=1)")
plt.xlabel("Valore")
plt.ylabel("Densità")
plt.legend()
plt.grid(True)
plt.show()

# 3. Stima parametro di forma con fit()
fitted_shape, loc, scale = gamma.fit(sample, floc=0)
print(f"Parametro stimato - shape: {fitted_shape:.4f}, scale: {scale:.4f}")

# 4. CDF stimata con i parametri stimati
cdf = gamma.cdf(x, a=fitted_shape, loc=0, scale=scale)

plt.figure(figsize=(10, 5))
plt.plot(x, cdf, label='CDF stimata', color='green')
plt.title("Funzione di Distribuzione Cumulativa (CDF)")
plt.xlabel("Valore")
plt.ylabel("Probabilità cumulata")
plt.grid(True)
plt.legend()
plt.show()

# 5. Calcola la varianza del campione
varianza = np.var(sample)
print(f"Varianza del campione: {varianza:.4f}")
```
