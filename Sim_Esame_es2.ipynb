Consideriamo un semplice processo di random walk (cammino aleatorio) in una dimensione:
a ogni passo temporale, un camminatore si sposta a destra (+1) o a sinistra (-1) con uguale probabilit√†. Supponiamo che il camminatore parta sempre da x=0.

1. Scrivi una funzione che simula il cammino per n passi ed N camminatori e restituisce due array. Il primo rappresenta la media della posizione ad ogni passo degli N camminatori ed il secondo la deviazione standard della posizione ad ogni passo degli N camminatori.

2. Fai prima il plot della media delle posizione degli N camminatori per 100 passi 

3. Fai il plot con anche la devizione standard.  La deviazione standard diminuisce all'aumentare del campione? Che andamento ha la deviazione standard rispeto l tempo/passi?

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Funzione che simula N camminatori per n passi
def random_walk_1D(n_passi, n_camminatori):
    # Genera array con valori -1 o +1 (movimenti), forma (n_camminatori, n_passi)
    passi = np.random.choice([-1, 1], size=(n_camminatori, n_passi))
    
    # Calcola la posizione cumulativa di ogni camminatore (asse tempo)
    posizioni = np.cumsum(passi, axis=1)
    
    # Inserisce uno zero all'inizio di ogni cammino (x=0)
    posizioni = np.insert(posizioni, 0, 0, axis=1)
    
    # Media e deviazione standard per ogni passo
    media_posizione = np.mean(posizioni, axis=0)
    std_posizione = np.std(posizioni, axis=0)

    return media_posizione, std_posizione

# Parametri
n_passi = 100
n_camminatori = 1000

# Simulazione
media, std = random_walk_1D(n_passi, n_camminatori)
x = np.arange(n_passi + 1)

# 2. Plot della media delle posizioni
plt.figure(figsize=(10, 5))
plt.plot(x, media, label='Media delle posizioni')
plt.title('Media delle posizioni dei camminatori nel tempo')
plt.xlabel('Passi')
plt.ylabel('Media posizione')
plt.grid(True)
plt.legend()
plt.show()

# 3. Plot con deviazione standard
plt.figure(figsize=(10, 5))
plt.plot(x, media, label='Media', color='blue')
plt.fill_between(x, media - std, media + std, color='lightblue', alpha=0.5, label='Deviazione standard')
plt.title('Media ¬± Deviazione standard delle posizioni')
plt.xlabel('Passi')
plt.ylabel('Posizione')
plt.grid(True)
plt.legend()
plt.show()
```
